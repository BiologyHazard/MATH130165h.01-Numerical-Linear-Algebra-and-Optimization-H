```python
import numpy as np
import matplotlib.pyplot as plt


iter_num = 1024

xs = []
x = 1/2
for i in range(iter_num):
    xs.append(x)
    x_new = x * (1 - x)
    x = x_new

fig, ax = plt.subplots()
ax.set_title('$x_{k+1} = x_k (1 - x_k)$')
ax.set_yscale('log')
ax.plot(xs, label='$x_k$')
ax.set_xlabel('Iteration')
ax.legend()
plt.show()
fig.savefig('../problem_7_plot.png', dpi=300)
```
